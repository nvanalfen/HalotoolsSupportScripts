```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
m_thresh = 10**np.array([9.5,10.0,10.5])
mu_cen = [0.5451, 0.6846, 0.7797]
mu_sat = [0.0300, 0.1390, 0.3345]

fig = plt.figure(figsize=(6,6))
p1, = plt.plot(m_thresh, mu_cen, 'o', ms=12)
p2, = plt.plot(m_thresh, mu_sat, 'x', ms=12)
plt.xscale('log')
plt.xlim([10**9,10**11.0])
plt.ylim([-0.25,1])
plt.ylabel(r'$\mu~[{\rm alignment~strength}]$', fontsize=20)
plt.xlabel(r'Stellar Mass Threshold $M_{*}$', fontsize=20)
plt.legend([p1,p2],['centrals','satellites'],frameon=True, fontsize=15, loc='upper left')

plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.tight_layout()
plt.ylim([-0.05,0.9])

plt.show()

fig.savefig("Alignment_Strength_progression.pdf", dpi=500)
```
