### $\chi^2_{dof}$

```python
cov_1xi = np.load( "data\\TNG300-1_99-sample_1_xi_cov_14_5x5x5.npy" )
cov_2xi = np.load( "data\\TNG300-1_99-sample_2_xi_cov_14_3x3x3.npy" )
cov_3xi = np.load( "data\\TNG300-1_99-sample_3_xi_cov_14_5x5x5.npy" )

cov_1ed = np.load( "data\\TNG300-1_99-sample_1_ed_cov_14_5x5x5.npy" )
cov_2ed = np.load( "data\\TNG300-1_99-sample_2_ed_cov_14_5x5x5.npy" )
cov_3ed = np.load( "data\\TNG300-1_99-sample_3_ed_cov_14_5x5x5.npy" )

cov_1ee = np.load( "data\\TNG300-1_99-sample_1_ee_cov_14_5x5x5.npy" )
cov_2ee = np.load( "data\\TNG300-1_99-sample_2_ee_cov_14_5x5x5.npy" )
cov_3ee = np.load( "data\\TNG300-1_99-sample_3_ee_cov_14_5x5x5.npy" )
```

```python
def chi_sq_dof(data_A, data_B, cov, dof=0, full_cov=True):
    
    diff = data_A - data_B
    
    if full_cov:
        inv_cov = np.linalg.inv(cov)
        return 0.5 * ( np.dot( diff, np.dot( inv_cov, diff ) ) / ( len(data_A) - dof ) )
    
    # assume cov is just err if not full_cov
    return np.sum( diff**2 / cov**2 ) / ( len(data_A) - dof )
```

```python
print( "Sample 1 Xi: ", chi_sq_dof( result_1_xi["xi"], model_1_xi["xi"], cov_1xi, 2 ) )
print( "Sample 2 Xi: ", chi_sq_dof( result_2_xi["xi"], model_2_xi["xi"], cov_2xi, 2 ) )
print( "Sample 3 Xi: ", chi_sq_dof( result_3_xi["xi"], model_3_xi["xi"], cov_3xi, 2 ) )
```

```python
print( "Sample 1 Xi: ", chi_sq_dof( result_1_xi["xi"], model_1_xi["xi"], result_1_xi["err"], 2, full_cov=False ) )
print( "Sample 2 Xi: ", chi_sq_dof( result_2_xi["xi"], model_2_xi["xi"], result_2_xi["err"], 2, full_cov=False ) )
print( "Sample 3 Xi: ", chi_sq_dof( result_3_xi["xi"], model_3_xi["xi"], result_3_xi["err"], 2, full_cov=False ) )
```

```python
print( "Sample 1 ED: ", chi_sq_dof( result_1_ed["ed"], model_1_ed["ed"], cov_1ed, 2 ) )
print( "Sample 2 ED: ", chi_sq_dof( result_2_ed["ed"], model_2_ed["ed"], cov_2ed, 2 ) )
print( "Sample 3 ED: ", chi_sq_dof( result_3_ed["ed"], model_3_ed["ed"], cov_3ed, 2 ) )
```
